```python
##nodes for use in a linked list
##two attributes: a value and a pointer
##get/set functions for each attribute
##print_method that prints out a node

class node:
    def __init__(self, value, prev_node = None, next_node = None):
        self.val = value
        self.prev = None
        self.next = None
        if prev_node is not None:
            self.prev = prev_node
        if next_node is not None:
            self.next = next_node
    def get_val(self):
        return self.val
    def get_prev(self):
        return self.prev
    def get_next(self):
        return self.next
    def set_val(self, value):
        self.val = value
    def set_prev(self, other_node):
        self.prev = other_node
    def set_next(self, other_node):
        self.next = other_node

    def print_method(self):
        print "Node value is: "+str(self.val)
        print "Prev node is:"+str(self.prev)
        print "Next node is:"+str(self.next)
```
